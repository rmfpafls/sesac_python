```python
def ploynomial(x, n): 
    y = x**n
    return y

x = np.linspace(-4,4, 100) #-4부터 4까지 100개의 데이터를 만들어 줘라  

y = ploynomial(x,1) # linear 
y2 = ploynomial(x,2)
y3 = ploynomial(x,3)
y4 = ploynomial(x,4)

plt.plot(x,y, label = "y=x")
plt.plot(x, y2, label = "y = x^2")
plt.plot(x, y3, label="y=x^3")
plt.plot(x, y4, label="y=x^4")

plt.title("Polynomial curve")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.legend() #라벨 출력하기 
plt.xlim(-4, 4) #그래프의 범위 보이기 
plt.ylim(-2,2)
plt.grid(linestyle = ":")
plt.show()
```

```python
# x 값 범위 설정 (-2에서 2까지 0.1 간격으로 값 생성)
x = np.arange(-2.0, 2.0, 0.1)

# 지수 함수 y = e^x 계산
y = 2**(x)
y1 = 2**(x-1)

# 그래프 그리기
plt.plot(x, y, label = "2^x")
plt.plot(x, y1, label = "2^(x-1)")

# 그래프에 제목과 축 라벨 설정
plt.title("Exponential Function: y = 2^(x-2)")
plt.vlines(0, 0, 16, color = "b") # 가운데 선 긋기 
plt.xlabel("x")
plt.ylabel("y")

# 그리드 추가 (선택사항)
plt.grid(True)

# 그래프 보여주기
plt.show()
```

# projector vetor 

```python
def projection_vec(base_vec, proj_vec):# a벡터가 base_vec, proj_vec가 b벡터임
    scale = np.inner(base_vec, proj_vec) / np.linalg.norm(base_vec)**2 #절댓값이 norm임
    p_vec = scale*base_vec
    plt.quiver(0, 0, base_vec[0], base_vec[1], angles = "xy",
                scale_units = "xy", color = 'r', scale = 1, label = "base_vetor")
    plt.quiver(0, 0, proj_vec[0], proj_vec[1], angles = "xy",
                scale_units = "xy", color = 'g', scale = 1, label = "Target_vetor")
    plt.quiver(0, 0, p_vec[0], p_vec[1], angles = "xy",
                scale_units = "xy", color = 'b', scale = 1, label = "projected_vetor")

    plt.xlim(-8, 8)
    plt.ylim(-8, 8)
    plt.grid(linestyle = ":")
    plt.legend()
    plt.show()

projection_vec(np.array([-2, 1]), np.array([-6, 4]))
    
```
