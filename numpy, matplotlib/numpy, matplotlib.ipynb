```python
def ploynomial(x, n): 
    y = x**n
    return y

x = np.linspace(-4,4, 100) #-4 4 100     

y = ploynomial(x,1) # linear 
y2 = ploynomial(x,2)
y3 = ploynomial(x,3)
y4 = ploynomial(x,4)

plt.plot(x,y, label = "y=x")
plt.plot(x, y2, label = "y = x^2")
plt.plot(x, y3, label="y=x^3")
plt.plot(x, y4, label="y=x^4")

plt.title("Polynomial curve")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.legend() #  
plt.xlim(-4, 4) #   
plt.ylim(-2,2)
plt.grid(linestyle = ":")
plt.show()
```

# projector vetor 

```python
def projection_vec(base_vec, proj_vec):# a base_vec, proj_vec b
    scale = np.inner(base_vec, proj_vec) / np.linalg.norm(base_vec)**2 # norm
    p_vec = scale*base_vec
    plt.quiver(0, 0, base_vec[0], base_vec[1], angles = "xy",
                scale_units = "xy", color = 'r', scale = 1, label = "base_vetor")
    plt.quiver(0, 0, proj_vec[0], proj_vec[1], angles = "xy",
                scale_units = "xy", color = 'g', scale = 1, label = "Target_vetor")
    plt.quiver(0, 0, p_vec[0], p_vec[1], angles = "xy",
                scale_units = "xy", color = 'b', scale = 1, label = "projected_vetor")

    plt.xlim(-8, 8)
    plt.ylim(-8, 8)
    plt.grid(linestyle = ":")
    plt.legend()
    plt.show()

projection_vec(np.array([-2, 1]), np.array([-6, 4]))
    
```
