```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. 그래프 그리기

```python
def ploynomal(x, n): 
	y = x**n
	return y

x = np.linspace(-4, 4, 100) 

y = ploynomal(x,1) # linear
y2 = ploynomal(x,2)

plt.plot(x, y, label = "y = x")
plt.plot(x, y2, label = "y = x^2")
plt.title("Polynomial curve")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.legend() #라벨 출력하기
plt.xlim(-4,4) # 그래프의 범위 보이기
plt.ylim(-10,10)
plt.grid(linestyle = ":")
plt.show()
```

# 3. numpy


```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)

sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

# for문 이용해서 그래프 3개 그리기
y_list = [sin_x, cos_x, tan_x]
y_label = ["sin_x", "cos_x", "tan_x"]
color_list = ['b', 'g', 'r']

fig, ax = plt.subplots(3, 1, figsize = (12, 10)) #3행 1열 

for i in range(3): 
    ax[i].plot(x, y_list[i], label = y_label[i], c = color_list[i])
    ax[i].set_ylim(-1,1)
    ax[i].grid(linestyle = ":")
    ax[i].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi","0","pi", "2pi"])

plt.show()
```

```python
# (1) y - b = sin(x - a)의 그래프는 y = sin(x)의 그래프를 x축으로 a, y축으로 b만큼 평행 이동한 것이다.
# (1) y = sin(x - a) + b의 그래프는 y = sin(x)의 그래프를 x축으로 a, y축으로 b만큼 평행 이동한 것이다.

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize = (12,5))
plt.plot(x, y, label = "sin(x)")
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ["-2pi", "-pi", "0", "pi", "2pi"])
plt.grid(linestyle = ":")
plt.legend()
plt.show()

```

# 4. 벡터와 행렬

```python
v = np.array([1,1])
w = np.array([-2,-1])

plt.quiver(0,0, v[0], v[1], angles = "xy", scale_units = "xy", scale = 1, color = "r")
plt.quiver(0,0, w[0], w[1], angles = "xy", scale_units = "xy", scale = 1, color = "b")
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.grid(linestyle = ":")
plt.show()
```
